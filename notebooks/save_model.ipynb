```python
# Load Data & Imports
import pandas as pd
import joblib

df = pd.read_csv("../data/processed/clean_telco_churn.csv")

X = df.drop(columns=["Churn"])
y = df["Churn"]
```

```python
# Rebuild the Final Model
from sklearn.ensemble import RandomForestClassifier

final_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=20,
    min_samples_leaf=10,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

final_model.fit(X, y)
```

```python
# Save the Model
joblib.dump(final_model, "../models/churn_random_forest_model.pkl")
```

```python
# Verify Reload Works
loaded_model = joblib.load("../models/churn_random_forest_model.pkl")

sample_prediction = loaded_model.predict(X.iloc[:5])
sample_prediction
```
